# EDA Titanic - Machine Learning from Disaster

## 1. Định nghĩa vấn đề (Define Problem)
+ **Mô tả**:
    + Bộ dữ liệu Titanic: Machine Learning from Disaster chứa thông tin về hành khách trên con tàu Titanic nổi tiếng bị đắm năm 1912.
    + Mục tiêu là dự đoán khả năng sống sót của hành khách dựa trên các đặc trưng cá nhân như tuổi, giới tính, hạng vé, giá vé, số người thân đi cùng,...
    + Đây là một bài toán phân loại nhị phân (binary classification), trong đó biến mục tiêu Survived có giá trị 0 (không sống sót) hoặc 1 (sống sót).
+ **Dữ liệu vào**:
    + PassengerId: Id định 
    + Pclass: Hạng vé (1 = hạng nhất, 2 = hạng nhì, 3 = hạng ba)
    + Name: Tên khách hàng
    + Sex: Giới tính của hành khách
    + Age: Tuổi của hành khách 
    + SibSp: Số anh/chị/em hoặc vợ/chồng đi cùng
    + Parch: Số cha mẹ hoặc con cái đi cùng
    + Fare: Giá vé (bằng bảng Anh)
    + Ticket: Mã vé
    + Cabin: Số Cabin
    + Embarked: Cảng lên tàu (C = Cherbourg, Q = Queenstown, S = Southampton)

+ **Kết quả**: Survived: 0 = tử vong, 1 = Sống sót

#### (5) **Mối tương quan giữa các tính chất (Correlations)**

Sự tương quan (correlation) đề cập đến mối quan hệ giữa hai biến và cách chúng có thể có hoặc không cùng nhau thay đổi.

Phương pháp phổ biến nhất để tính toán tương quan là Pearson's Correlation Coeficient, giả định có một phân phối chuẩn của các thuộc tính liên quan. Tương quan -1 hoặc 1 cho thấy mối tương quan âm hoặc dương đầy đủ tương ứng. Trong khi giá trị 0 hiển thị không tương quan ở tất cả.

$$
r = \frac{\sum_{i=1}^n{\left(x_i - \hat{x}\right)\left(y_i - \hat{y}\right)}}{\sqrt{\sum_{i=1}^n{\left(x_i - \hat{x}\right)^2}\sum_{i=1}^n{\left(y_i - \hat{y}\right)^2}}}
$$

Một số thuật toán học máy như hồi quy tuyến tính và logistic có hiệu suất kém nếu có các thuộc tính tương quan cao trong tập dữ liệu của bạn.

Như vậy, thật sự cần thiết để xem xét tất cả các mối tương quan theo cặp của các thuộc tính trong tập dữ liệu.

```python
plt.figure(figsize=(12, 6));
plt.imshow(plt.imread("corr1.png")), plt.axis("off"), plt.show();
plt.imshow(plt.imread("corr2.png")), plt.axis("off"), plt.show();
```

```python
correlations = df_dataset[["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]].corr()
display.display(correlations)
```

**Nhận xét**:
Một số cặp biến có mức độ tương quan đáng chú ý:
+ (Survived, Pclass) = -0.338 → Mối tương quan âm vừa phải, thể hiện hành khách ở hạng vé cao (Pclass nhỏ) có tỷ lệ sống sót cao hơn.
+ (Survived, Fare) = 0.257 → Mối tương quan dương nhẹ, gợi ý rằng giá vé cao (tức hạng vé tốt) thường đi kèm khả năng sống sót cao hơn.
+ (SibSp, Parch) = 0.415 → Cho thấy mối quan hệ dương giữa số anh chị em/vợ chồng và số cha mẹ/con cái đi cùng — tức những hành khách đi cùng gia đình thường có cả hai chỉ số này cao hơn.
+ Các cặp biến còn lại như (Survived, Age) hay (Fare, Age) có hệ số tương quan nhỏ (< 0.1), cho thấy mối liên hệ yếu hoặc gần như không đáng kể giữa tuổi, giá vé và khả năng sống sót.

##### Box and whisker plots

+ So sánh các trung vị (median) tương ứng của mỗi ô hộp (box plot). Nếu đường trung vị của một ô hộp nằm bên ngoài ô của một ô hộp so sánh, thì có thể có sự khác biệt giữa hai nhóm.
+ So sánh chiều dài hộp để kiểm tra cách dữ liệu được phân tán giữa mỗi mẫu. Hộp càng dài thì dữ liệu càng phân tán. Dữ liệu càng nhỏ càng ít bị phân tán.
+ Một ngoại lệ (outlier) được định nghĩa là một điểm dữ liệu nằm bên ngoài phần rìa (whiskers) của ô hộp.
+ Kiểm tra hướng lệch của dữ liệu (cân đối, các phần tử tập trung trái, phải).
    + Median ở giữa hộp và râu (whiskers) ở hai bên như nhau thì phân bố là đối xứng.
    + Median ở gần đáy hộp hơn và nếu râu ngắn hơn ở đầu dưới của hộp, thì phân phối là lệch dương (lệch phải).
    + Median ở gần đầu hộp hơn và nếu râu ngắn hơn ở đầu trên của hộp, thì phân bố bị lệch âm (lệch trái).

```python
plt.figure(figsize=(12, 6)), plt.imshow(plt.imread("boxplots.png")), plt.axis("off"), plt.show();
```

```python
cols = ["Age", "Fare", "SibSp", "Parch"]
df_dataset[cols].plot(kind="box", subplots=True, layout=(2, 2),
                              sharex=False, sharey=False, figsize=(12, 6))
plt.show()
```

**Nhận xét**:
+ Age: Phân bố tương đối cân bằng, trung vị (median) khoảng 28–30 tuổi. Một số giá trị ngoại lai nằm trên 60–80 tuổi, thể hiện sự hiện diện của hành khách lớn tuổi nhưng số lượng ít.
+ Fare: Có phân phối rất lệch phải (right-skewed) với nhiều giá trị ngoại lệ ở mức cao. Phần lớn hành khách trả giá vé thấp (<100), trong khi một số ít hành khách hạng cao có vé lên tới hơn 500.
+ SibSp (số anh/chị/em hoặc vợ/chồng đi cùng): Đa số hành khách đi một mình, chỉ một vài trường hợp có từ 3–8 người đi cùng.
